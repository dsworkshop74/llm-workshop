```python
%%writefile app.py
import streamlit as st
import os
from docling.document_converter import DocumentConverter
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_core.documents import Document
from langchain_chroma import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain import hub
from langchain_core.runnables import RunnablePassthrough
import tempfile
import time

# Page configuration
st.set_page_config(
    page_title="📚 Document Q&A Assistant",
    page_icon="🤖",
    layout="wide",
    initial_sidebar_state="expanded"
)

# Custom CSS for better styling
st.markdown("""
<style>
    .main-header {
        font-size: 3rem;
        font-weight: bold;
        text-align: center;
        background: linear-gradient(90deg, #FF6B6B, #4ECDC4, #45B7D1);
        background-clip: text;
        -webkit-background-clip: text;
        -webkit-text-fill-color: transparent;
        margin-bottom: 2rem;
    }

    .sub-header {
        text-align: center;
        color: #666;
        font-size: 1.2rem;
        margin-bottom: 3rem;
    }

    .stTextInput > div > div > input {
        border-radius: 10px;
        border: 2px solid #e1e5e9;
        padding: 10px;
    }

    .stTextArea > div > div > textarea {
        border-radius: 10px;
        border: 2px solid #e1e5e9;
    }

    .success-box {
        padding: 1rem;
        border-radius: 10px;
        background-color: #d4edda;
        border: 1px solid #c3e6cb;
        color: #155724;
        margin: 1rem 0;
    }

    .info-box {
        padding: 1rem;
        border-radius: 10px;
        background-color: #d1ecf1;
        border: 1px solid #bee5eb;
        color: #0c5460;
        margin: 1rem 0;
    }

    .warning-box {
        padding: 1rem;
        border-radius: 10px;
        background-color: #fff3cd;
        border: 1px solid #ffeaa7;
        color: #856404;
        margin: 1rem 0;
    }

    .sidebar-content {
        background-color: #f8f9fa;
        padding: 1rem;
        border-radius: 10px;
        margin: 1rem 0;
    }
</style>
""", unsafe_allow_html=True)

# Initialize session state
if 'retriever' not in st.session_state:
    st.session_state.retriever = None
if 'document_processed' not in st.session_state:
    st.session_state.document_processed = False
if 'conversation_history' not in st.session_state:
    st.session_state.conversation_history = []

@st.cache_resource
def load_models():
    """Load and cache the embedding model and LLM"""
    try:
        embeddings_model = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
        llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash-lite")
        return embeddings_model, llm
    except Exception as e:
        st.error(f"Error loading models: {str(e)}")
        return None, None

def get_markdown(source):
    """Convert document from source to markdown"""
    try:
        converter = DocumentConverter()
        result = converter.convert(source)
        text = result.document.export_to_markdown()
        return text
    except Exception as e:
        st.error(f"Error converting document: {str(e)}")
        return None

def create_langchain_doc(context_text):
    """Split text and create LangChain documents"""
    try:
        text_splitter = RecursiveCharacterTextSplitter(
            chunk_size=1000,
            chunk_overlap=50,
            length_function=len,
        )
        documents = text_splitter.split_text(context_text)
        langchain_formated_doc = []
        for i in range(len(documents)):
            langchain_formated_doc.append(Document(page_content=documents[i]))
        return langchain_formated_doc
    except Exception as e:
        st.error(f"Error creating documents: {str(e)}")
        return None

def create_retriever(langchain_formated_doc, embeddings_model):
    """Create vector database and retriever"""
    try:
        db = Chroma.from_documents(langchain_formated_doc, embeddings_model)
        retriever = db.as_retriever()
        return retriever
    except Exception as e:
        st.error(f"Error creating retriever: {str(e)}")
        return None

def fetch_source_docs(question, retriever):
    """Fetch relevant documents for the question"""
    try:
        sim_vector = retriever.invoke(question)
        return sim_vector
    except Exception as e:
        st.error(f"Error fetching documents: {str(e)}")
        return None

def process_document(source_url):
    """Process the document and create retriever"""
    with st.spinner("🔄 Processing document..."):
        # Convert document to markdown
        context_text = get_markdown(source_url)
        if context_text is None:
            return False

        # Create LangChain documents
        langchain_docs = create_langchain_doc(context_text)
        if langchain_docs is None:
            return False

        # Load models
        embeddings_model, llm = load_models()
        if embeddings_model is None or llm is None:
            return False

        # Create retriever
        retriever = create_retriever(langchain_docs, embeddings_model)
        if retriever is None:
            return False

        # Store in session state
        st.session_state.retriever = retriever
        st.session_state.llm = llm
        st.session_state.document_processed = True

        return True

def answer_question(question):
    """Answer question using RAG pipeline"""
    try:
        # Load prompt template
        prompt = hub.pull("rlm/rag-prompt")

        # Create RAG pipeline
        def get_context(state):
            return fetch_source_docs(state, st.session_state.retriever)

        rag_pipeline = {
            "context": get_context,
            "question": RunnablePassthrough()
        } | prompt | st.session_state.llm

        # Get response
        response = rag_pipeline.invoke(question)
        return response.content
    except Exception as e:
        st.error(f"Error answering question: {str(e)}")
        return None

# Main UI
st.markdown('<h1 class="main-header">📚 Document Q&A Assistant</h1>', unsafe_allow_html=True)
st.markdown('<p class="sub-header">Upload any document via URL and ask questions about it!</p>', unsafe_allow_html=True)

# Sidebar
with st.sidebar:
    st.markdown("### 🛠️ Setup")

    # API Key input
    google_api_key = st.text_input(
        "Google API Key",
        type="password",
        help="Enter your Google API key for Gemini"
    )

    if google_api_key:
        os.environ["GOOGLE_API_KEY"] = google_api_key

    st.markdown("---")

    # Document source input
    st.markdown("### 📄 Document Source")
    source_url = st.text_input(
        "Document URL",
        placeholder="https://example.com/document.pdf",
        help="Enter the URL of the document you want to analyze"
    )

    process_btn = st.button("🔄 Process Document", type="primary", use_container_width=True)

    if process_btn and source_url:
        if not google_api_key:
            st.error("Please enter your Google API key first!")
        else:
            success = process_document(source_url)
            if success:
                st.success("✅ Document processed successfully!")
            else:
                st.error("❌ Failed to process document")

    # Display processing status
    if st.session_state.document_processed:
        st.markdown('<div class="success-box">✅ Document ready for questions!</div>', unsafe_allow_html=True)

    st.markdown("---")

    # Instructions
    st.markdown("### 📋 Instructions")
    st.markdown("""
    1. **Enter your Google API Key** (required for Gemini)
    2. **Paste the document URL** (PDF, Word, etc.)
    3. **Click 'Process Document'** and wait
    4. **Ask questions** about the document content
    5. **View answers** with relevant context
    """)

    # Clear conversation button
    if st.button("🗑️ Clear Conversation", use_container_width=True):
        st.session_state.conversation_history = []
        st.experimental_rerun()

# Main content area
if not st.session_state.document_processed:
    # Welcome message
    st.markdown("""
    <div class="info-box">
        <h3>🚀 Welcome to Document Q&A Assistant!</h3>
        <p>This powerful tool allows you to:</p>
        <ul>
            <li>📥 Process documents from any web URL</li>
            <li>🤖 Ask intelligent questions about the content</li>
            <li>💡 Get accurate answers with relevant context</li>
            <li>📊 Maintain conversation history</li>
        </ul>
        <p><strong>Get started by entering your Google API key and document URL in the sidebar!</strong></p>
    </div>
    """, unsafe_allow_html=True)

    # Sample URLs for testing
    st.markdown("### 🔗 Sample URLs to try:")
    col1, col2 = st.columns(2)

    with col1:
        st.code("https://en.wikipedia.org/wiki/Sachin_Tendulkar")
        st.code("https://en.wikipedia.org/wiki/Machine_learning")

    with col2:
        st.code("https://en.wikipedia.org/wiki/Python_(programming_language)")
        st.code("https://en.wikipedia.org/wiki/Artificial_intelligence")

else:
    # Question input area
    st.markdown("### 💬 Ask a Question")

    col1, col2 = st.columns([4, 1])

    with col1:
        question = st.text_input(
            "Your Question",
            placeholder="What would you like to know about the document?",
            key="question_input"
        )

    with col2:
        ask_btn = st.button("🔍 Ask", type="primary", use_container_width=True)

    # Process question
    if (ask_btn and question) or (question and st.session_state.get('last_question') != question):
        if question.strip():
            st.session_state.last_question = question

            with st.spinner("🤔 Thinking..."):
                answer = answer_question(question)

                if answer:
                    # Add to conversation history
                    st.session_state.conversation_history.append({
                        "question": question,
                        "answer": answer,
                        "timestamp": time.strftime("%Y-%m-%d %H:%M:%S")
                    })

    # Display conversation history
    if st.session_state.conversation_history:
        st.markdown("### 📝 Conversation History")

        for i, conv in enumerate(reversed(st.session_state.conversation_history)):
            with st.expander(f"Q: {conv['question'][:100]}{'...' if len(conv['question']) > 100 else ''}", expanded=(i==0)):
                st.markdown(f"**🕒 Time:** {conv['timestamp']}")
                st.markdown(f"**❓ Question:** {conv['question']}")
                st.markdown(f"**💡 Answer:**")
                st.write(conv['answer'])

    # Quick question suggestions
    if not st.session_state.conversation_history:
        st.markdown("### 💡 Quick Question Suggestions")
        col1, col2, col3 = st.columns(3)

        suggestions = [
            "What is the main topic?",
            "Give me a summary",
            "What are the key points?",
            "Tell me about the author",
            "What are the conclusions?",
            "Explain the methodology"
        ]

        for i, suggestion in enumerate(suggestions):
            col = [col1, col2, col3][i % 3]
            with col:
                if st.button(suggestion, key=f"suggestion_{i}", use_container_width=True):
                    st.session_state.question_input = suggestion
                    st.experimental_rerun()

# Footer
st.markdown("---")
st.markdown("""
<div style="text-align: center; color: #666; padding: 2rem;">
    <p>🤖 Powered by Docling, LangChain, ChromaDB, and Google Gemini</p>
    <p>Built with ❤️ using Streamlit</p>
</div>
""", unsafe_allow_html=True)
```
