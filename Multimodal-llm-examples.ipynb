# Text only

```python
# pip install transformers
from transformers import AutoModelForCausalLM, AutoTokenizer
checkpoint = "HuggingFaceTB/SmolLM-1.7B-Instruct"

device = "cuda" # for GPU usage or "cpu" for CPU usage
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = AutoModelForCausalLM.from_pretrained(checkpoint).to(device)
```

```python
messages = [{"role": "user", "content": "What is the capital of France."}]
input_text=tokenizer.apply_chat_template(messages, tokenize=False)
print(input_text)
```

```python
inputs = tokenizer.encode(input_text, return_tensors="pt").to(device)
inputs
```

```python
outputs = model.generate(inputs, max_new_tokens=50, temperature=0.2, top_p=0.9)
outputs
```

```python
print(tokenizer.decode(outputs[0]))
```

```python
from transformers import AutoProcessor, AutoModelForImageTextToText
import torch

model_path = "HuggingFaceTB/SmolVLM2-2.2B-Instruct"
processor = AutoProcessor.from_pretrained(model_path)
model = AutoModelForImageTextToText.from_pretrained(
    model_path,
    torch_dtype=torch.bfloat16,
).to("cuda")
```

```python
!wget "https://upload.wikimedia.org/wikipedia/commons/b/b6/Image_created_with_a_mobile_phone.png"
```

```python
from PIL import Image
image_path = '/content/Image_created_with_a_mobile_phone.png'
image = Image.open(image_path)
image.resize((250, 250))
```

```python
image.resize((250, 250)).save(image_path)
```

```python
messages = [
    {
        "role": "user",
        "content": [
            {"type": "text", "text": "Describe the Image?"},
            {"type": "image", "path": image_path}
        ]
    }
]

messages
```

```python
inputs = processor.apply_chat_template(
    messages,
    add_generation_prompt=True,
    tokenize=True,
    return_dict=True,
    return_tensors="pt",
).to(model.device, dtype=torch.bfloat16)

inputs
```

```python
inputs.keys()
```

--> pixel_values

A tensor (e.g., torch.Tensor) containing the processed image data.

These are the normalized pixel values (after resizing, cropping, converting to tensors, etc.) for each image in your chat messages.

The processor uses its internal image handler to handle formats like URLs or local paths, preparing them for the model


--> pixel_attention_mask

Indicates which image patches are padding and which contain real image data.

For example, if images are resized or padded to a fixed shape, this mask ensures the model doesn’t “focus” on padded areas.

It parallels attention_mask in NLP, but for visual inputs
stackoverflow.com
.

--> input_ids

A tensor of token IDs representing your text.

These IDs come from the tokenizer interpreting your chat prompts (including system messages, user messages, assistant messages) according to the model’s chat template

--> attention_mask

A binary tensor (0s and 1s) marking which text tokens should be attended to and which are padding.

This ensures the model ignores padding tokens during processing (common in Transformers)

```python
generated_ids = model.generate(**inputs, max_new_tokens=64)
generated_ids
```

```python
generated_texts = processor.batch_decode(
    generated_ids,
    skip_special_tokens=True,
)
print(generated_texts[0])
```
